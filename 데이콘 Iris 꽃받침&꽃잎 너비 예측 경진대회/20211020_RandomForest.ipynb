```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv("iris_train.csv")
test = pd.read_csv("iris_test.csv")
```

```python
from ngboost import NGBRegressor
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import *
from sklearn.linear_model import Ridge, Lasso

from sklearn.model_selection import KFold, StratifiedKFold, LeaveOneOut
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder
```

```python
lo = LeaveOneOut()
kf = KFold(n_splits = 5, random_state = 42, shuffle = True)
```

```python
train.groupby('species')['sepal width (cm)'].mean()
```

```python
train.groupby('species')['petal width (cm)'].mean()
```

```python
train['sep_w'] = train['species'].map({'setosa' : 3, 'versicolor' : 1, 'virginica' : 2})
test['sep_w'] = test['species'].map({'setosa' : 3, 'versicolor' : 1, 'virginica' : 2})
```

```python
train['pet_w'] = train['species'].map({'setosa' : 1, 'versicolor' : 2, 'virginica' : 3})
test['pet_w'] = test['species'].map({'setosa' : 1, 'versicolor' : 2, 'virginica' : 3})
```

```python
X = train.drop(['species', 'sepal width (cm)', 'petal width (cm)'], axis = 1)
y1 = train.iloc[:, 4]
y2 = train.iloc[:, 5]
```

```python
target = test[X.columns]
```

```python
rf_pred1 = np.zeros((target.shape[0]))
for i, idx in enumerate(zip(lo.split(X, y1))) :
    print(f'{i + 1} FOLD Start!')
    tr_x, tr_y = X.iloc[idx[0][0]], y1.iloc[idx[0][0]]
    val_x, val_y = X.iloc[idx[0][1]], y1.iloc[idx[0][1]]
    
    rf = RandomForestRegressor(random_state = 42, n_estimators = 100, criterion = 'mae', max_depth = 4)
    
    rf.fit(tr_x, tr_y)
    val_pred = rf.predict(val_x)
    val_mae = mean_absolute_error(val_y, val_pred)
    print(f'VAL MAE = {val_mae}')
    fold_pred = rf.predict(target) / 75
    
    rf_pred1 += fold_pred
    
    print('\n')
```

```python
rf_pred2 = np.zeros((target.shape[0]))
for i, idx in enumerate(zip(lo.split(X, y2))) :
    print(f'{i + 1} FOLD Start!')
    tr_x, tr_y = X.iloc[idx[0][0]], y2.iloc[idx[0][0]]
    val_x, val_y = X.iloc[idx[0][1]], y2.iloc[idx[0][1]]
    
    rf = RandomForestRegressor(random_state = 42, n_estimators = 100, criterion = 'mae', max_depth = 4)
    
    rf.fit(tr_x, tr_y)
    val_pred = rf.predict(val_x)
    val_mae = mean_absolute_error(val_y, val_pred)
    print(f'VAL MAE = {val_mae}')
    fold_pred = rf.predict(target) / 75
    
    rf_pred2 += fold_pred
    
    print('\n')
```

```python
submission = pd.read_csv("sample_submission.csv")
```

```python
submission.iloc[:, 1] = rf_pred1
submission.iloc[:, 2] = rf_pred2
```

```python
submission.to_csv('1020_rf_1st.csv', index = False)
```
