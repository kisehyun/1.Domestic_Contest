### Loss Function Definition

```python
def pb_loss(true, pred, q) :
    
    L_list = []
    
    for i in range(len(true)) :
        
        if true.iloc[i] >= pred[i] :
            L = true.iloc[i] - pred[i]
            L *= q
            
        else :
            L1 = pred[i] - true.iloc[i]
            L2 = 1 - q
            L = L1 * L2
            
        L_list.append(L)
        
    return np.mean(L_list)
```

```python
def make_data() :
    ### 1일 후 데이터
    train_after_1 = train.loc[train.Day <= 1093]
    train_after_1['TARGET1'] = list(train.loc[train.Day >= 1].TARGET)
    
    ### 2일 후 데이터
    train_after_2 = train.loc[train.Day <= 1092]
    train_after_2['TARGET2'] = list(train.loc[train.Day >= 2].TARGET)
    
    ### after 1 day data 
    a_1day = test.loc[test.Day != 6].drop(['Day', 'Minute', 'month', 'sep_day','DNI', 'RH'], axis = 1)
    
    ### after 2 day data 
    a_2day = test.query('Day not in [5, 6]').drop(['Day', 'Minute','month','sep_day','DNI', 'RH'], axis = 1)
    
    ### after 1 day train data
    X_1 = train_after_1.drop(['Day','Minute','new_day', 'TARGET1','month','DNI', 'RH'], axis = 1)
    y_1 = train_after_1.TARGET1
    
    ### after 2 days train data
    X_2 = train_after_2.drop(['Day','Minute', 'new_day', 'TARGET2','month','DNI', 'RH'], axis = 1)
    y_2 = train_after_2.TARGET2
    
    ### actual values
    real_1 = test.loc[test.Day != 0].TARGET
    real_2 = test.query('Day not in [0, 1]').TARGET
    
    target = test.loc[test.Day == 6].drop(['Day', 'sep_day', 'Minute','month','DNI', 'RH'], axis = 1)
    
    return a_1day, a_2day, X_1, y_1, X_2, y_2, real_1, real_2, target
```

```python
def iter_lgbm(q):
    params = {'max_depth' : [3,4,5,6,7], 'alpha' : np.arange(q, q + 0.08, 0.01), 'n_estimators' : [50,100, 150, 200]}
    
    sc_df = pd.DataFrame({'max_depth' : [], 'alpha' : [], 'n_estimators' : [], 'best_loss' : []})
    i = -1
    for m in params['max_depth'] :
        for a in params['alpha'] :
            for n in params['n_estimators'] :
                i += 1
                lgbm = LGBMRegressor(random_state = 101, metric = 'quantile', objective = 'quantile', max_depth = m, n_estimators = n, alpha = a)
                lgbm.fit(X_1, y_1, eval_set = [(a_1day, real_1)], early_stopping_rounds = 1000, verbose = 0)
                best_loss = float(str(lgbm.best_score_['valid_0']).split(',')[-1][1:-3])
                sc_df.loc[i] = [m, a, n, best_loss]
    
    best_comb = sc_df.sort_values(by = 'best_loss', ascending = False).tail(1)
    use_params = {'max_depth' : int(best_comb.max_depth.values[0]), 'n_estimatros' : int(best_comb.n_estimators.values[0]), 'alpha' : best_comb.alpha.values[0]}
    
    lgbm = LGBMRegressor(random_state = 101, **use_params, metric = 'quantile', objective = 'quantile')
    lgbm.fit(X_1, y_1, eval_set = [(a_1day, real_1)], early_stopping_rounds = 1000, verbose = 0)
    best_loss = float(str(lgbm.best_score_['valid_0']).split(',')[-1][1:-3])
    print(f'best loss는 {best_loss}\n')
    lgbm_pred = [0 if x < 0 else x for x in lgbm.predict(target)]
    
    return lgbm_pred
```

```python
def lgbm_opt(learning_rate, n_estimators, max_depth, subsample, colsample_bytree, num_leaves, min_child_weight, alpha):
    params = {
        'learning_rate': learning_rate,
        'n_estimators' : int(round(n_estimators)),
        'max_depth' : int(round(max_depth)),
        'subsample': subsample,
        'colsample_bytree' : colsample_bytree,
        'num_leaves' : int(round(num_leaves)),
        'min_child_weight' : min_child_weight,
        'n_jobs' : -1,
        'alpha' : alpha
    }
    

    score = []
    for i, (t, a) in enumerate(zip(t_list, a_list)) :
        lgbm = LGBMRegressor(**params, random_state = 1231)
        lgbm.fit(X_1, y_1, eval_set = [(t, a)], early_stopping_rounds = 1000, verbose = 0)
        best_loss = -float(str(lgbm.best_score_['valid_0']).split(',')[-1][1:-3])
        score.append(best_loss)
        
    return np.mean(score) 
```
