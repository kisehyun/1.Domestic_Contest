```python
X = train.loc[:, 'X_01':'X_56']
y = train.loc[:, 'Y_01':]
```

```python
target = test[X.columns]
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import make_scorer, mean_squared_error
```

```python
def lg_nrmse(gt, preds):
    # 각 Y Feature별 NRMSE 총합
    # Y_01 ~ Y_08 까지 20% 가중치 부여
    all_nrmse = []
    for idx in range(14): # ignore 'ID'
        rmse = mean_squared_error(gt.iloc[:,idx], preds[:,idx], squared = False)
        nrmse = rmse / np.mean(np.abs(gt.iloc[:,idx]))
        all_nrmse.append(nrmse)
    score = 1.2 * np.sum(all_nrmse[:8]) + 1.0 * np.sum(all_nrmse[8:15])
    return score
```

```python
kf = KFold(n_splits = 10, random_state = 42, shuffle = True)
```

```python
NRMSES = []
cb_pred = np.zeros((target.shape[0], 14))
for i, idx in enumerate(kf.split(X, y)) :
    tr_x, tr_y = X.loc[idx[0]], y.loc[idx[0]]
    val_x, val_y = X.loc[idx[1]], y.loc[idx[1]]
    
    cb = CatBoostRegressor(random_state = 42, max_depth = 5, n_estimators = 10000, learning_rate = 0.03, objective = 'MultiRMSE')
    cb.fit(tr_x, tr_y, eval_set = [(tr_x, tr_y), (val_x, val_y)], early_stopping_rounds = 500, verbose = 2000)
    
    val_pred = cb.predict(val_x)
    NRMSE = lg_nrmse(val_y, val_pred)
    print(f"{i + 1} Fold NRMSE = {NRMSE}")
    NRMSES.append(NRMSE)
    
    fold_pred = cb.predict(target) / kf.n_splits
    cb_pred += fold_pred
```

```python
np.mean(NRMSES)
```

```python
submission = pd.read_csv('sample_submission.csv')
```

```python
submission.iloc[:, 1:] = cb_pred
```

```python
submission.to_csv('1st.csv', index = False)
```
