```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from catboost import CatBoostClassifier, Pool
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, LeaveOneOut, cross_val_score
```

```python
skf = StratifiedKFold(n_splits = 10, random_state = 42, shuffle = True)
```

```python
X = train.drop('target', axis = 1)
y = train.target
```

```python
target = test.copy()
```

```python
X.shape, target.shape
```

```python
X.columns
```

```python
target.columns
```

```python
cat_cols = ['cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']
```

```python
cb_pred = np.zeros(target.shape[0])
cb_f1 = 0
for tr_idx, val_idx in skf.split(X, y) :
    tr_x, tr_y = X.iloc[tr_idx], y.iloc[tr_idx]
    val_x, val_y = X.iloc[val_idx], y.iloc[val_idx]
    
    tr_data = Pool(data = tr_x, label = tr_y, cat_features = cat_cols)
    val_data = Pool(data = val_x, label = val_y, cat_features = cat_cols)
    
    cb = CatBoostClassifier(random_state = 42, verbose = 0, n_estimators = 1000, max_depth = 3)
    cb.fit(tr_data, eval_set = val_data, early_stopping_rounds = 200, verbose = 250)
    
    val_pred = cb.predict(val_x)
    val_f1 = f1_score(val_y, val_pred) / 10
    cb_f1 += val_f1
    
    fold_pred = cb.predict_proba(target)[:, 1] / 10
    cb_pred += fold_pred
print(f'\n{np.mean(cb_f1)}')
```

```python
final_pred = [1 if p >= 0.5 else 0 for p in cb_pred]
```

```python
submission.target = final_pred
```

```python
submission.to_csv("20211215_1st.csv", index = False)
```
